```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import csv
```

```python
df = pd.read_csv('ob_report_2023.csv')
print(df.head())
print(df.columns)

```

```python
font1 = {'family': 'sans-serif', 'color':'navy', 'size':18}
font2 = {'family': 'monospace', 'color':'darkgreen', 'size':14}
np.random.seed(200)
fig, ax = plt.subplots(figsize = (5,3))
df['Outbreak Setting'].value_counts().plot(kind='bar', width = 0.5)
ax.set_title('Number of Outbreaks in Healthcare Insitutions', fontdict=font1, pad=20)
ax.set_ylabel('Frequency', fontdict= font2)
ax.set_xticklabels(labels= ['LTCH', 'Retirement Home', 'Hopsital-Chronic Care', 'Hospital-Acute Care', 'Hospital-Psychiatric', 'Transitional Care'], rotation=45, ha= 'right')
ax.set_xlabel('Outbreak Setting', fontdict= font2)
fig.tight_layout
```
