```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# set random seed for reproducibility
np.random.seed(61)
x = np.arange(50)
y = np.random.randint(0,100,50)

# making a scatterplot using x,y values specified above witha figure size of 5,3
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y)
```

```python
# making a bar graph using the same x,y axis and points
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y)
```

```python
# making a plot using the x,y points as specified above
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
# making a histogram using the x,y points and figure size specified above
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
# setting figure axis and figure size
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

# setting axes titles
ax.set_title('Total growth over time')
# setting y-axis label
ax.set_ylabel('Total growth')
# setting x-axis label
ax.set_xlabel('Years since start')
```

```python
# setting font1 with a specific font, color and size
font1 = {'family':'sans-serif', 'color' : 'blue', 'size': 20}
# setting font2 with a specific font, color and size
font2 = {'family':'monospace', 'color' : 'green', 'size': 14}
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
# changing axes title using font1 and location allignment left
ax.set_title('Total Growth Over Time', fontdict = font1, loc = 'left')
# changing y-axis label using font2
ax.set_ylabel('Total growth', fontdict = font2)
# changing x-axis label using font2
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# changing axes styles further with the marker type, color, style and width 
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        color = 'indigo',
        linestyle = '--',
        linewidth = '2')
```

```python
# changing axes styles further with the marker type, color, style and width 
# changing axes styles further with the marker edge color and marker face color

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,
        y,
        marker = '*',
        markersize= 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
# adding grid lines to our axis object
ax.grid(axis= 'y')
```
